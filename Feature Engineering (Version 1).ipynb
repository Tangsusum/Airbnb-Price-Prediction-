# Feature Engineering 

## Binary Features

```python
# Drop useless binary features 
cols_to_drop = [ 'requires_license',
                 'is_business_travel_ready',
                 'require_guest_profile_picture',
                 'require_guest_phone_verification']
data = data.drop(cols_to_drop, axis = 1)
```

## Categorical Features

### Cancellation Policy

```python
# One hot encoding
to_encode = ['cancellation_policy']
for j in to_encode:
    u = []
    for i in data[j]:
        if i not in u:
            u.append(i)
    one_hot = pd.get_dummies(data[j])
    data = data.join(one_hot)
```

```python
# Group all others into property_type_others
data['cancellation_type_others'] = data.apply(lambda row: 1-(row.strict_14_with_grace_period + row.flexible + row.moderate), axis = 1)

# Drop all other columns 
property_type = data['cancellation_policy'].unique()
clean_property_type = []
for p in property_type:
    if p != 'strict_14_with_grace_period' and p != 'flexible' and p != 'moderate':
        clean_property_type.append(p)
        
data = data.drop(columns = clean_property_type)
data = data.drop(columns = 'cancellation_policy')
```

### Neighbourhood cleansed

```python
# One hot encoding 
to_encode = ['neighbourhood_cleansed']
for j in to_encode:
    u = []
    for i in data[j]:
        if i not in u:
            u.append(i)
    one_hot = pd.get_dummies(data[j])
    data = data.join(one_hot)
    
# Drop original column 
data = data.drop('neighbourhood_cleansed', axis = 1)

# Make sure all names does not contain space 
data.colums = data.columns.str.replace("[()]/", "_")
data.columns = [c.replace(' ', '_') for c in data.columns]
```

### Property Type

```python
# One hot encoding
to_encode = ['property_type']
for j in to_encode:
    u = []
    for i in data[j]:
        if i not in u:
            u.append(i)
    one_hot = pd.get_dummies(data[j])
    data = data.join(one_hot)
    
```

```python
# Group serviced apartment with apartment 
data.Apartment = data.Apartment + data['Serviced apartment']

# Group House, townhouse and condominum together 
data.House = data.House + data.Townhouse + data.Condominium

# Group all others into property_type_others
data['property_type_others'] = data.apply(lambda row: 1-(row.Apartment + row.House), axis = 1)

# Drop all other columns 
property_type = data['property_type'].unique()
clean_property_type = []
for p in property_type:
    if p != 'House' and p != 'Apartment':
        clean_property_type.append(p)
        
data = data.drop(columns = clean_property_type)
data = data.drop(columns = 'property_type')
```

### Room Type

```python
# One hot encoding
to_encode = ['room_type']
for j in to_encode:
    u = []
    for i in data[j]:
        if i not in u:
            u.append(i)
    one_hot = pd.get_dummies(data[j])
    data = data.join(one_hot)
    
# Make sure all names does not contain space 
data.colums = data.columns.str.replace("[()]/", "_")
data.columns = [c.replace(' ', '_') for c in data.columns]
```

```python
# Shared room and Hotel room into room_type_others
data['room_type_others'] = data.apply(lambda row: row.Shared_room + row.Hotel_room, axis = 1)

# Drop all other columns         
data = data.drop(columns = ['room_type','Shared_room','Hotel_room'])
```

### Bed type

```python
# Drop because it doesn't have much explanatory power
cols_to_drop = [ 'bed_type']
data = data.drop(cols_to_drop, axis = 1)
```

### Host Response Time

```python
# One hot encoding 
to_encode = ['host_response_time']
for j in to_encode:
    u = []
    for i in data[j]:
        if i not in u:
            u.append(i)
    one_hot = pd.get_dummies(data[j])
    data = data.join(one_hot)
    
# Drop original column 
data = data.drop('host_response_time', axis = 1)

# Make sure all names does not contain space 
data.colums = data.columns.str.replace("[()]/", "_")
data.columns = [c.replace(' ', '_') for c in data.columns]
```

### Host Verification

```python
from numpy import asarray
from numpy import savetxt
from numpy import loadtxt
# One hot for host_verification
x = data['host_verifications']
x = [item.strip('[').strip(']').split(",") for item in x]
x = [subitem.lstrip().strip("'") for item in x for subitem in item]

from collections import Counter
count = Counter(x)
sorted_keys_v = [k for k, v in sorted(count.items(), key=lambda item: item[1])]

# remove irregular key 
sorted_keys_v.remove('')

# save sort_keys
savetxt('sorted_keys_v.csv',sorted_keys_v, delimiter = ',', fmt = "%s")

# include all 
selected = sorted_keys_v[-5:]
var = np.zeros((len(data), 5))
for i, info in enumerate(data['host_verifications']):
    for j, s in enumerate(selected):
        if s in info:
            var[i, j] = 1;
            
df = {}
for j, s in enumerate(selected):
    df[s+'_dummy'] = var[:, j]
    
df = pd.DataFrame(df)
del data['host_verifications']
data = data.join(df)
```

### Ammenities

```python
# One hot for amenities
x = data['amenities']
x = [item.strip("{").strip("}").replace('"', '').split(",") for item in x]
x = [word for item in x for word in item]

from collections import Counter
count = Counter(x)
sorted_keys_a = [k for k, v in sorted(count.items(), key=lambda item: item[1])]

# remove irregular key 
sorted_keys_a.remove('')

# save keys
for i,w in enumerate(sorted_keys_a):
    sorted_keys_a[i] = w.replace(" ","_")
savetxt('sorted_keys_a.csv',sorted_keys_a, delimiter = ',',fmt = "%s")

# Pick all
selected = sorted_keys_a[-5:]
var = np.zeros((len(data), 5))
for i, info in enumerate(data['amenities']):
    for j, s in enumerate(selected):
        if s in info:
            var[i, j] = 1;
            
new_df = {}
for j, s in enumerate(selected):
    new_df[s+'_dummy'] = var[:, j]
    
new_df = pd.DataFrame(new_df)
del data['amenities']
data = data.join(new_df)
```

## Time/date features

```python
# Use today as the baseline
# Replace the date by the number of days

today = datetime.date.today()
idx = data.columns.get_loc('host_since')

date_format = "%Y-%m-%d"
for i,date in enumerate(data.host_since):
    d = datetime.datetime.strptime(date,date_format).date()
    diff = today - d
    data.iloc[i,idx] = diff.days
```

```python
# Use today as the baseline
# Replace the date by the number of days
# There are missing values in first and last review
# ignore if detected 

col = ['first_review','last_review']

for c in col:
    today = datetime.date.today()
    idx = data.columns.get_loc(c)

    date_format = "%Y-%m-%d"

    for i,date in enumerate(data[c]):
        try:
            d = datetime.datetime.strptime(date,date_format).date()
            diff = today - d
            data.iloc[i,idx] = diff.days
        except:
            print('NAN')
```

```python
# Replace null values with 0 
to_fill = ['first_review',
          'last_review']
for i in to_fill:
    data[i] = data[i].fillna(0)
```

## Standardise numerical predictors

```python
# Normalise numerical predictors
to_norm = ['host_acceptance_rate',
            'security_deposit_perc',
            'review_scores_value',
            'review_scores_checkin',
            'review_scores_location',
            'review_scores_accuracy',
            'review_scores_communication',
            'review_scores_cleanliness',
            'review_scores_rating',
            'cleaning_fee_perc',
            'host_response_rate',
            'reviews_per_month',
            'bathrooms',
            'bedrooms',
            'beds',
            'host_since',
            'first_review',
            'last_review',
            'host_listings_count',
            'host_total_listings_count',
            'calculated_host_listings_count',
            'calculated_host_listings_count_entire_homes',
            'calculated_host_listings_count_private_rooms',
            'calculated_host_listings_count_shared_rooms',
            'number_of_reviews_ltm',
            'number_of_reviews',
            'minimum_nights',
            'maximum_nights',
            'minimum_minimum_nights',
            'minimum_maximum_nights',
            'maximum_maximum_nights',
            'maximum_minimum_nights',
            'minimum_nights_avg_ntm',
            'maximum_nights_avg_ntm',
            'guests_included',
            'accommodates',
            'availability_30',
            'availability_60',
            'availability_90',
            'availability_365']

for c in to_norm:
    mean = data[c].mean()
    maximum = data[c].max()
    minimum = data[c].min()
    std = data[c].std()
    data[c] = (data[c] - mean)/std
```

```python
removed_outliers = data['bedrooms'].between(data['bedrooms'].quantile(.05), data['bedrooms'].quantile(.95))
index_names = data[~removed_outliers].index
data.drop(index_names, inplace=True)
removed_outliers = data['beds'].between(data['beds'].quantile(.05), data['beds'].quantile(.95))
index_names = data[~removed_outliers].index
removed_outliers = data['bathrooms'].between(data['bathrooms'].quantile(.05), data['bathrooms'].quantile(.95))
index_names = data[~removed_outliers].index
data.drop(index_names, inplace=True)
```
